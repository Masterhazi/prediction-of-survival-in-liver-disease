```python
def bilirubin_score(bilirubin):
    if bilirubin < 2:
        return 1
    elif 2 <= bilirubin <=3:
        return 2
    else:
        return 3

def albumin_score(albumin):
    if albumin>3.5:
        return 1
    elif 2.8<=albumin<=3.5:
        return 2
    else:
        return 3

def ascites_score(ascites):
    if ascites == 'N':
        return 1
    else:
        return 3

def prothrombin_score(prothrombin):
    if prothrombin<4:
        return 1
    elif 4<=prothrombin<=6:
        return 2
    else:
        return 3
```

```python
df['Bilirubin_Score'] = df['Bilirubin'].apply(bilirubin_score)
df['Albumin_Score'] = df['Albumin'].apply(albumin_score)
df['Ascites_Score'] = df['Ascites'].apply(ascites_score)
df['Prothrombin_Score'] = df['Prothrombin'].apply(prothrombin_score)

```

```python
df['Child_Pugh_Score'] = df[['Bilirubin_Score', 'Albumin_Score', 'Ascites_Score', 'Prothrombin_Score']].sum(axis = 1)
```

```python
test['Bilirubin_Score'] = test['Bilirubin'].apply(bilirubin_score)
test['Albumin_Score'] = test['Albumin'].apply(albumin_score)
test['Ascites_Score'] = test['Ascites'].apply(ascites_score)
test['Prothrombin_Score'] = test['Prothrombin'].apply(prothrombin_score)
test['Child_Pugh_Score'] = test[['Bilirubin_Score', 'Albumin_Score', 'Ascites_Score', 'Prothrombin_Score']].sum(axis = 1)
```

```python
from quicklearnkit import train_test_split

X = df.drop('target', axis = 1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle= True, stratify=y, random_state = 42)
```

```python
import numpy as np

print('Train dist:')
unique, counts = np.unique(y_train, return_counts = True)

print("\nTest distribution:")
unique, counts = np.unique(y_test, return_counts = True)

print(dict(zip(unique, counts / len(y_test))))
```

```python
X_train_encoded = pd.get_dummies(X_train, drop_first = True).astype('int')
X_test_encoded = pd.get_dummies(X_test, drop_first = True).astype('int')

X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join = 'left', axis = 1, fill_value=0)
```

```python
X_test_encoded
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators= 400, max_depth = None, min_samples_split= 2, class_weight= 'balanced', random_state = 42, n_jobs = -1)

rf.fit(X_train_encoded, y_train)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

y_pred_proba = rf.predict_proba(X_test_encoded)[:,1]
y_pred = rf.predict(X_test_encoded)

```

```python
X_test_encoded
```

```python
y_test
```

```python
print("ROC AUC:", roc_auc_score(y_test, y_pred_proba))
print('\nclassification_report', classification_report(y_test, y_pred))
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1][:15]

plt.figure(figsize= [10,6])
plt.title('Top 15 features according to importance')
plt.bar(range(len(indices)), importances[indices])
plt.xticks(range(len(indices)), X_train_encoded.columns[indices], rotation = 90)
plt.show()
```
